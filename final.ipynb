```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("usgs_data/earthquakes_unificado.csv")

df.head()
```

```python
df.isnull().sum()
```

```python
df["time"] = pd.to_datetime(df["time"], unit="ms", errors="coerce")
df["updated"] = pd.to_datetime(df["updated"], unit="ms", errors="coerce")
```

```python
# Ver Relaciones entre las columnas de mag. depth, gap y rms
df[['mag','depth','gap','rms']].corr()
```

```python
df['time'].isna().sum()
```

```python
for col in df.columns:
    total_null = df[col].isnull().sum()
    resultado = total_null / len(df) *100
    print(f"Columna {col}: {total_null} valores nulos | porcentaje de nulos {resultado:.2f}%\n")
```

```python
df['magType'].isnull().sum()
```

```python
df['magType'].value_counts()
```

```python
# En la base de datos ya viene un valor Unknown, usaremos ese valor para los valores NaN
df['magType'] = df['magType'].fillna('Unknown')
```

```python
df['gap'].mean()
```

```python
df['gap'].isnull().sum()
```

```python
# Eliminar columna dmin
df.drop(columns=['dmin'], inplace=True)
```

```python
# Crear columna de decada a partir de time
df['decada'] = df['time'].dt.year // 10 * 10
```
